```python
#b. Understanding how many new customers acquired every month
df_final['Month'] = df_final.order_purchase_timestamp.dt.month
df_final['Year'] = df_final.order_purchase_timestamp.dt.year
display(df_final.groupby(['Year','Month'])[['customer_unique_id']].nunique())
```

```python
##Understand the retention of customers on month on month basis

df_final['Delivered'] =df_final.order_status[df_final.order_status == 'delivered']
df_final['Month'] = pd.DataFrame(df_final.order_purchase_timestamp.dt.month)
df_final['Year'] = pd.DataFrame(df_final.order_purchase_timestamp.dt.year)

retention_cust = pd.pivot_table(data = df_final, index = ['Month','Delivered'],columns = 'Year',values ='customer_id', aggfunc = 'count')
retention_cust
```

```python
## How the revenues from existing/new customers on month on month b


df_final['Revenue'] = df_final.payment_value-df_final.freight_value
revenue = pd.pivot_table(data = df_final, index = ['Month'],columns = 'Year',values ='Revenue', aggfunc = np.sum)
revenue
```

```python
##Understand the trends/seasonality of sales, quantity by category, location, month,week, day, time, channel, payment method etc

## sales by loaction

df_final.groupby(['geolocation_state','Year'],as_index = False).sum().pivot('geolocation_state','Year').fillna(0)
```

```python
##Popular Products by month, seller, state, category.

df_final.groupby(['product_category_name','Year'],as_index = False).count().pivot('product_category_name','Year').fillna(0).sort_values(by ='product_category_name', ascending = False)
```

```python
###Popular categories by state, month

df_final.groupby(['product_category_name','geolocation_state'],as_index = False).count().pivot('product_category_name','geolocation_state').fillna(0).sort_values(by ='product_category_name', ascending = False)
```
